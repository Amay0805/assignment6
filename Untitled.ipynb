```python
"""Problem 1: Bank Account Create a class representing a bank account with attributes like account number, 
account holder name, and balance. Implement methods to deposit and withdraw money from the account."""

class bankaccount:
    def __init__(self,acc_no,holder,balance):
        self.account_no=acc_no
        self.holder=holder
        self.balance=balance
    def withdraw(self,amount):
        self.balance=self.balance-amount
        return f"You have successfully withdrawn: {amount}\nNow your account balance is: {self.balance}"
    def deposit(self,amount):
        self.balance=self.balance+amount
        return f"You have successfully deposited: {amount}\nNow your account balance is: {self.balance}"
    
a=bankaccount(123456789,"xyz",5000)
print(f"The Account Number of the account is {a.account_no}\nAccount Holder is {a.holder}\nAccount balance is {a.balance}")
print(a.withdraw(1000))
print(a.deposit(3000))
```

```python
# Problem 3 
# Vehicle Rental 
# Create a class representing a vehicle rental system.
# Implement methods to rent a vehicle, return a vehicle, and display available vehicles.

class Rental_Vehicles:
    def __init__(self):
        self.list=["Renault KWID","Tata Tiago","Maruti Suzuki Swift","New Maruti Dzire","Maruti Vitara Brezza"
              ,"Hyundai Elite i20","Honda WR-V","Hyundai Verna","Hyundai Creta","Jeep Compass"]
    def available_vehicles(self):
        a=""
        for i in range(len(self.list)):
            a+=f"\n{i+1}) {self.list[i]}"
        return f"The avialable vehicles are:{a}"
    def rentvehicle(self,vehicle):
        self.list.remove(vehicle)
        a=""
        for i in range(len(self.list)):
            a+=f"\n{i+1}) {self.list[i]}"
        return f"You have selected {vehicle} for rent."
    def vehiclereturn(self,vehicle):
        self.list.append(vehicle)
        a=""
        for i in range(len(self.list)):
            a+=f"\n{i+1}) {self.list[i]}"
        return f"Thank You for returning our vehicle {vehicle}."
a=Rental_Vehicles()
print(a.available_vehicles())
print()
print(a.rentvehicle("Jeep Compass"))
print()
print(a.vehiclereturn("Jeep Compass"))
print()
print(a.available_vehicles())
```

```python
#Problem 5
#Product Inventory
#Create classes representing a product and an inventory system
#Implement methods to add products to the inventory, update product quantity, and display available products

class Product_Inventory:
    def __init__(self):
        self.product_inventory = {"Coffee Maker": 50,"Smartphone Case": 25,"Wireless Earbuds": 40,"Laptop Bag": 30,"Fitness Tracker": 15,
        "Portable Bluetooth Speaker": 20,"Coffee Grinder": 12,"External Hard Drive": 18,"Reusable Water Bottle": 35,"Desk Organizer": 28}
    def available_products(self):
        a=""
        count=0
        for product in self.product_inventory:
            count+=1
            a+=f"\n{count}) {product}: {self.product_inventory[product]} products available"
        return f"The available products are:{a}"
    def add_products(self,productname,quantity):
        if productname in self.product_inventory:
            self.product_inventory[productname]+=quantity
        else:
            self.product_inventory[productname]=quantity
        return f"Thank you for successfully adding the product {productname} by {quantity} pieces"
    def update_quantity(self,productname,quantity):
        if productname in self.product_inventory:
            self.product_inventory[productname]=quantity
            return f"You have successfully updated the quantity of {productname} to {quantity}"
        else:
            return f"This method is used for updating the quantity of the products that are present in the available products.\n{productname} not available in the available products"
            
obj=Product_Inventory()
print(obj.available_products())
print()
print(obj.add_products("Smartphone",11))
print()
print(obj.add_products("Smartphone Case",11))
print()
print(obj.available_products())
print()
print(obj.update_quantity("Smartphone Case",25))
print()
print(obj.available_products())
```

```python
#Problem 7
#Student Management
#Create a class representing a student with attributes like student ID, name, and grades
#Implement methods to calculate the average grade and display student details

class student:
    def __init__(self,student_id,name,grades):
        self.student_id=student_id
        self.name=name
        self.grades=grades
    def average_grade(self):
        self.sum=0
        for i in self.grades:
            self.sum+=i
        self.average=self.sum/len(self.grades)
        return self.average
    def display_student_student_details(self):
        a=""
        a+=f"The student id is: {self.student_id}\n"
        a+=f"The student name is: {self.name}\n"
        a+=f"The student name is: {self.name}\n"
        a+=f"The average marks of the student is: {self.average}"
        return a
obj=student(123456789,"abc",[99,88,88,81,80])
print(obj.average_grade())
print(obj.display_student_student_details())
```

```python
#ToDo List
#Create a class representing a ToDo list with attributes like tasks and due dates
#Implement methods to add tasks, mark tasks as completed, and display pending tasks

class ToDoList:
    def __init__(self):
        self.tasks=[]
    def add_task(self,tasks,due_dates):
        self.tasks.append({'task':tasks,'due dates':due_dates,'completed':False})
    def mark_task_completed(self,task_index):
        if 0<=task_index<len(self.tasks):
            self.tasks[task_index]['completed']=True
            print(f"Task {self.tasks[task_index]['task']} marked as completed ")
        else:
            print("invalid task index")
    def display_pending_tasks(self):
        pending_tasks = [task for task in self.tasks if not task["completed"]]
        if pending_tasks:
            print("Pending Tasks:")
            for idx, task in enumerate(pending_tasks, start=1):
                print(f"{idx}. Task: {task['task']}   Due Date: {task['due dates']}")
        else:
            print("No pending tasks.")

todo_list = ToDoList()

todo_list.add_task("Complete project proposal", "2023-05-15")
todo_list.add_task("Buy groceries", "2023-05-20")
todo_list.add_task("Read a chapter of a book", "2023-05-18")

todo_list.display_pending_tasks()

todo_list.mark_task_completed(1)

todo_list.display_pending_tasks()
```
